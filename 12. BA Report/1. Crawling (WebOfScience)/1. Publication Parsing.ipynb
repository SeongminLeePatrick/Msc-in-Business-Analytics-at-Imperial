# Publication information on Web of Science

example link : http://apps.webofknowledge.com/InboundService.do?customersID=LinksAMR&mode=FullRecord&IsProductCode=Yes&product=WOS&Init=Yes&Func=Frame&DestFail=http%3A%2F%2Fwww.webofknowledge.com&action=retrieve&SrcApp=PARTNER_APP&SrcAuth=LinksAMR&SID=E4J7VrDZl7zNFz5RH3S&UT=WOS%3A000080526900002

To collect the information relevant to the publication, I visited each professor's imperial website, such as "http://www.imperial.ac.uk/people/professor_id/publications.html?limit=400." And among the publications, I just parsed the URLs linked with the Web of Science, which is the website to have organised information on each publication. Before starting the web-scraping in the Web of Science, I found out that there are some duplicate publications with different co-authors. Therefore, the primary key for each publication is the publications' URL on the Web of Science. Also, to avoid ID block, the four ids from my group have been used in a way that each log-in was executed by selecting one of the ids at random. Also, because the Web of Science server sometimes blocks the log-in on the IP address with the message "???en_US.Server.IDLimit???", the scraping action will be automatically aborted after five times trial. All in all, I could get the publication's information as below:

1. Title: the name of a publication
2. WOS_URL: This will be used as a primary key if there is a join.
3. co_author1: author ids linked with the author information
4. co_author2: The other type of the author name designated by the author
5. journal_type: Official journal name of the publisher
6. research_areas: Web of Science research Areas
7. WOS_cate: Web of Science research Category
8. address: The addresses of the authors including organisation, country, and city
9. organisation: Official organisation name
10. date: published date
11. cited_time: the number of being cited
12. num_of_ref: the number of the reference in a publication
13. abstract: abstract of a publication
14. keywords: keywords provided by authors
15. publisher: publisher address

## Convert csv file to pickle file due to the unexpected commas in the list

Such as co_author1/2, research_ares, WOS_cate, addreess, and organization columns, it is difficult to handle the list with the strings due to the punctuation. So, the literal_eval function helped to remove the punctuation in the list and keep the data in a right format to preprocess.

```python
read_data = pd.read_csv('./data/WOS_scraped_df_ver3.csv')

# pickle save error
for col in read_data.columns:
    try:
        read_data[col] = read_data[col].apply(lambda x: literal_eval(x))
    except:
        pass

read_data.to_pickle('./data/WOS_scraped_df_ver3.pkl')
```
