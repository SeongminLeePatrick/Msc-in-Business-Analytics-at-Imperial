# Reference on Web of Science
**Example link** : http://apps.webofknowledge.com/summary.do?product=WOS&parentProduct=WOS&search_mode=CitedRefList&parentQid=24&parentDoc=1&qid=25&SID=E4J7VrDZl7zNFz5RH3S&colName=WOS&page=1

Cited references in a publication could have a positive effect on increasing the cited time for the publication. This is because a reference in a publication could support the publication's credibility and reliability, which is an important factor to be cited. On top of that, an assumption that the higher cited time a publication has, the more references with the higher cited time are quoted. Therefore, I scraped the references' title, cited time, and year of the publication.

# Initial Crawling

```python
def reference_type():
    WOS_df = pd.read_pickle('WOS_scraped_df_ver2.pkl')
    WOS_df_set_URL = list(set(WOS_df.WOS_URL)) 
    try:
        ref_df = pd.read_csv('ref_info_version2.csv')
        last_work_url = list(ref_df.publication_url)[-1]
        startidx = WOS_df_set_URL.index(last_work_url) + 1
    except:
        ref_df = pd.DataFrame(columns = ['publication_url','citing_title', 'citing_year', 'ref_title','ref_year', 'ref_times_cited'])
        startidx = 0

    
    for url in WOS_df_set_URL[startidx:]:
        #print("Start! %d. Document URL: %s" %(url))
#         print(WOS_df.loc[ind, 'Title'])
        #Log-in on Web of Science
        #Select the id and pass at random
        ids = ['Maja.pavlovic17@imperial.ac.uk', 'vpapageorgiou23@gmail.com', 'markosh94@hotmail.com']
         #ids = ['imperialgroupk@gmail.com']
        selected_id = random.choice(ids)
        passwords = ['WFarmy100!', '#imperial17', '#imperial17']
         #passwords = ['#imperial17']
        following_pass = passwords[ids.index(selected_id)]
        #Log-in on Web of Science
        chrome_options = Options()
        chrome_options.add_argument('--lang=en-us')
        driver = webdriver.Chrome(chrome_options=chrome_options,executable_path=r"C:/Users/seong/Downloads/chrome/chromedriver")
        driver.get('https://login.webofknowledge.com/error/Error?PathInfo=%2F&Error=IPError')
        driver.implicitly_wait(10)
        driver.find_element_by_name('username').send_keys(selected_id) #email id
        driver.find_element_by_name('password').send_keys(following_pass) #password
        driver.implicitly_wait(10)
        driver.find_element_by_xpath('//*[@id="WoKerror"]/div/table[2]/tbody/tr/td[2]/form[1]/table/tbody/tr[3]/td[2]/a/button').click() #click
        driver.implicitly_wait(10)
        try:
            driver.find_element_by_xpath('//*[@id="WoKerror"]/div/table[2]/tbody/tr/td[2]/p/a[1]').click() #exception log-in
            driver.implicitly_wait(10)
        except:
            pass

        # Reaching the main page for the publication on Web of Science
        driver.get(url)
        
        try: # Reaching the reference DB on Web of Science
            driver.find_element_by_xpath('//*[@id="cited-refs-full-record"]/div[1]/h3/a').click()
            driver.implicitly_wait(20)
        except: #no link
            row_num = ref_df.shape[0]
            loc = np.where(WOS_df.WOS_URL == url)[0][0]
            ref_df.loc[row_num, 'publication_url'] = WOS_df.WOS_URL[loc]
            ref_df.loc[row_num, 'citing_title'] = WOS_df.Title[loc]
            ref_df.loc[row_num, 'citing_year'] = WOS_df.date[loc]
            ref_df.loc[row_num, 'ref_times_cited'] = 'no_reference'
            ref_df.loc[row_num, 'ref_year'] = 'no_reference'
            ref_df.loc[row_num, 'ref_title'] = 'no_reference'
            continue
            
        driver.implicitly_wait(10)    
        web_soup3 = BeautifulSoup(driver.page_source, "html.parser")
        end_page = int(web_soup3.select_one("span#pageCount.top").string)            
        loc = np.where(WOS_df.WOS_URL == url)[0][0]
        for page in range(end_page):
            
            if page == 0:
                for i in web_soup3.find_all("div", id=re.compile("^RECORD_")):
                    row_num = ref_df.shape[0]
                    ref_df.loc[row_num, 'publication_url'] = WOS_df.WOS_URL[loc]
                    ref_df.loc[row_num, 'citing_title'] = WOS_df.Title[loc]
                    ref_df.loc[row_num, 'citing_year'] = WOS_df.date[loc]
                    try:
                        ref_df.loc[row_num, 'ref_times_cited'] = int((i.select_one("div.search-results-data-cite > a").string).replace(",",""))
                    except:
                        ref_df.loc[row_num, 'ref_times_cited'] = np.nan
                        
                    try:
                        ref_df.loc[row_num, 'ref_year'] = i.find(text = re.compile(r'^Published:')).next_element.get_text().strip()
                    except:
                        ref_df.loc[row_num, 'ref_year'] = np.nan
                        
                    try:
                        ref_df.loc[row_num, 'ref_title'] = i.select_one("span.reference-title > value").string
                    except:
                        ref_df.loc[row_num, 'ref_title'] = np.nan

            else:
                driver.find_element_by_xpath('//*[@id="summary_navigation"]/nav/table/tbody/tr/td[3]/a/i').click()
                driver.implicitly_wait(20)
                web_soup3 = BeautifulSoup(driver.page_source, "html.parser")
                for i in web_soup3.find_all("div", id=re.compile("^RECORD_")):
                    row_num = ref_df.shape[0]
                    ref_df.loc[row_num, 'publication_url'] = WOS_df.WOS_URL[loc]
                    ref_df.loc[row_num, 'citing_title'] = WOS_df.Title[loc]
                    ref_df.loc[row_num, 'citing_year'] = WOS_df.date[loc]
                    try:
                        ref_df.loc[row_num, 'ref_times_cited'] = int((i.select_one("div.search-results-data-cite > a").string).replace(",",""))
                    except:
                        ref_df.loc[row_num, 'ref_times_cited'] = np.nan
                    
                    try:
                        ref_df.loc[row_num, 'ref_year'] = i.find(text = re.compile(r'^Published:')).next_element.get_text().strip()
                    except:
                        ref_df.loc[row_num, 'ref_year'] = np.nan
                    
                    try:
                        ref_df.loc[row_num, 'ref_title'] = i.select_one("span.reference-title > value").string
                    except:
                        ref_df.loc[row_num, 'ref_title'] = np.nan
         
        print("%sth Added with the title: %s" %(ref_df.citing_title.nunique(), WOS_df.WOS_URL[loc]))
        ref_df.to_csv('ref_info_version2.csv', index = None)
        driver.close()
    driver.quit()
```

```python
for idx, t in enumerate(ref_info.ref_title):
    if t == 'no_reference':
        ref_info.loc[idx, 'ref_num_count'] = 0
    else:
        ref_info.loc[idx, 'ref_num_count'] = 1
```

```python
wos = pd.read_pickle('./data/WOS_scraped_df_ver2.pkl')
df_tem = pd.DataFrame(ref_info.groupby('publication_url').sum()['ref_num_count'])
df_tem['publication_url'] = df_tem.index
merge_df = wos.merge(df_tem, how='left', left_on = 'WOS_URL', right_on = 'publication_url')
unmatched_index = np.where(merge_df.num_of_ref != merge_df.ref_num_count)[0] #error searching
```
